```python
from manim import *
```

```python
%%manim -v WARNING --disable_caching -qm Test

config.media_width = "100%"



class Test(LinearTransformationScene):
    def construct(self):

        t1 = MathTex(" M =  \\begin{bmatrix} 2 & 1 \\\ 1 & 2 \\end{bmatrix}   ").to_edge(DL).add_background_rectangle()
        self.add(t1)    
    
        self.add_title("Linear Transformations")
        unit_square = self.get_unit_square()
        self.add_transformable_mobject(unit_square)
        matrix = [[2,1],[1,2]]
        self.apply_matrix(matrix)
```

```python
%%manim -v WARNING --disable_caching -qm Test2



class Test2(Scene):
    def construct(self):
        com = ComplexPlane()
        com.add_coordinates()
        self.add(com)
```

```python
%%manim -v WARNING --disable_caching -qm Test3 

config.media_width = "100%"



class Test3(Scene):
    def construct(self):
        func = lambda pos: ((pos[0]*UR+pos[1]*LEFT)-pos)/3
        vector_field = ArrowVectorField(func)
        self.add(vector_field)
        stream_lines = StreamLines(func, color = BLUE)
        self.add(stream_lines)
        stream_lines.start_animation(warm_up=False, flow_speed=1.5)
        self.wait(stream_lines.virtual_time / stream_lines.flow_speed)
```
